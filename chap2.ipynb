```python
import numpy as np
import matplotlib.pyplot as plt
```

問題2.3

```python
dx = 0.1
dt = 0.05
c = 1

x = np.arange(0.0, 2*np.pi, dx)
nmax = len(x)

u = np.sin(x)
flux = np.zeros(u.shape)

step = 60
for t in range(step+1):
    flux = 1/2 * (c * u + c * np.roll(u, -1)) - 1/2 * abs(c) * (np.roll(u, -1) - u)
    u += -dt/dx * (flux - np.roll(flux, 1))

    #周期境界条件なので上だけでOK

    #出力
    if t % 20 == 0:
        plt.plot(x, u, marker='o', label=f't = {t}')

plt.grid()
plt.legend(loc='upper right')
plt.show()
```

問題2.4

```python
dx = 0.05
dt = 0.001
kappa = 1.0
sigma = 0.1
print(dt <= dx**2/2/kappa)

x = np.arange(-1.0, 1.0, dx)
nmax = len(x)

u = 1.0 / (np.sqrt(2.0 * np.pi) * sigma) * np.exp(-x**2 / (2.0 * sigma**2))
 
step = int(0.2 / dt)
for t in range(step+1):
    u += kappa * dt / dx**2 * (np.roll(u, -1) - 2.0 * u + np.roll(u, 1))

    #対称境界条件
    u[0] = u[3]
    u[1] = u[2]
    u[-1] = u[-4]
    u[-2] = u[-3]
    #u[0] = u[1]
    #u[-1] = u[-2]
    
    #出力
    if t % int(0.025 / dt) == 0:
        plt.plot(x, u, marker='o', label=f't = {t * dt :.3f}')

plt.grid()
plt.legend(loc='upper right')
plt.show()
```
